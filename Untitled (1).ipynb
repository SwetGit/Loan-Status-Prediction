4.Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

ax = sns.countplot(x='Education', hue='Loan_Status', data=data)

# Add count labels on bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', fontsize=10)

plt.title("Loan Status Distribution by Education")
plt.xlabel("Education")
plt.ylabel("Count")
plt.show()
```

```python
# Conclusion-
# Graduates apply more for loans compared to non-graduates (higher total bar height).
# Graduates have a higher number of approved loans (Y) than non-graduates.
# But loan approval rate is not drastically different between the two groups, suggesting that education alone may not strongly influence approval.
# Both education groups show more approved loans than denied ones, indicating an overall higher approval rate.

```

```python
#Converting Categorical Value to numerical Value as model understands numerical better
pd.set_option('future.no_silent_downcasting', True)

data.replace({
    "Married": {'No': 0, 'Yes': 1},
    "Gender": {'Female': 0, 'Male': 1},
    "Self_Employed": {'Yes': 1, 'No': 0},
    "Education": {'Graduate': 1, 'Not Graduate': 0},
    "Property_Area": {'Urban': 2, 'Rural': 0, 'Semiurban': 1}
}, inplace=True)
```

```python
type(data)
```

```python
#Separating features and Lables: x-->features ,y-->labels
x = data.drop(columns=['Loan_Status', 'Loan_ID'], axis=1)
y=data["Loan_Status"]
print(x)

```

```python
print(y)
```

```python
# Splitting Training Data and Test Data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
print(x.shape, x_train.shape,x_test.shape)
```

Training the model by svm

```python
classifier=svm.SVC(kernel='linear')
classifier.fit(x_train, y_train)
```

Evaluating Model

```python
x_train_prediction=classifier.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction, y_train)
print(training_data_accuracy)
```

#On test data

```python
x_test_prediction=classifier.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction, y_test)
print(test_data_accuracy)
```
